```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
from pylab import rcParams
from scipy.stats import f_oneway
from scipy.stats import ttest_ind
```

```python
%matplotlib inline
warnings.filterwarnings("ignore")
rcParams['figure.figsize'] = 18,12
rcParams['font.size'] = 60
sns.set()
np.random.seed(8)
```

```python
def plot_distribution(inp):
    plt.figure()
    ax = sns.distplot(inp)
    plt.axvline(np.mean(inp), color="k", linestyle="dashed", linewidth=5)
    _, max_ = plt.ylim()
    plt.text(
        inp.mean() + inp.mean() / 10,
        max_ - max_ / 10,
        "Mean: {:.2f}".format(inp.mean()),
    )
    return plt.figure
```

```python
#Mpumalanga
mpumalanga = duration_not_null[new_df2['province'] == "Mpumalanga"]
```

```python
mpum_duration = mpumalanga['duration']
```

```python
mpum_duration.mean()
```

```python
plot_distribution(mpum_duration)
```

```python
plt.figure()
ax1 = sns.distplot(mpum_duration)
ax2 = sns.distplot(free_duration)
plt.axvline(np.mean(mpum_duration), color='b', linestyle='dashed', linewidth=5)
plt.axvline(np.mean(free_duration), color='orange', linestyle='dashed', linewidth=5)
```

```python
def compare_2_groups(arr_1, arr_2, alpha, sample_size):
    stat, p = ttest_ind(arr_1, arr_2)
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    if p > alpha:
        print('Same distributions (fail to reject H0)')
    else:
        print('Different distributions (reject H0)')
```

```python
sample_size = 1000
df_Mpu_sampled = np.random.choice(mpum_duration, sample_size)
df_Fs_sampled = np.random.choice(free_duration, sample_size)
compare_2_groups(df_Mpu_sampled, df_Fs_sampled, 0.05, sample_size)
```
